```python
import pandas as pd
import numpy as np
from lightautoml.automl.presets.tabular_presets import TabularAutoML, TabularUtilizedAutoML
from lightautoml.tasks import Task
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss, roc_auc_score, roc_curve, precision_score, recall_score, f1_score, precision_recall_curve
```

```python
df = pd.read_csv('housing.csv')
```

```python
RANDOM_STATE = 1000
```

```python
df_train, df_test = train_test_split(df, test_size=0.2, random_state=RANDOM_STATE)
```

```python
y_train = df_train['MEDV']
y_test = df_test['MEDV']
```

```python
task = Task(name='reg')
roles = {'target': 'MEDV'}
automl = TabularAutoML(task=task, timeout=3600, memory_limit=10)
oof_pred = automl.fit_predict(df_train, roles=roles)
```

```python
pred = automl.predict(df.drop('MEDV', axis = 1)).data
```

```python
df['pred'] = pred
```
