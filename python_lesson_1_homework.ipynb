```python
# Сколько вакансий для аналатика данных?
a=0
for _, i in enumerate(vacs):
    if i['vactitle'].find('Аналитик данных') !=-1 or i['vactitle'].find('Data analyst') !=-1:
        a +=1        
print(a)
    
```

```python
# Сколько вакансий с позициями на которых вы работаете?
n=0
for _, i in enumerate(vacs):
    if i['vactitle'].find('Риск-менеджер') !=-1 or i['vactitle'].find('Risk manager') !=-1:
        n +=1        
print(n)
```

```python
# Сколько вакансий, которые вам нравятся?
t=0
for _, i in enumerate(vacs):
    if i['vactitle'].find('Разработчик') !=-1 or i['vactitle'].find('Аналитик данных') !=-1 or i['vactitle'].find('Data analyst') !=-1:
        t +=1        
print(t)
```

```python
# Сколько вакансий для аналитика данных с использованием Python?
b=0
for _, i in enumerate(vacs):
    if i['vactitle'].find('Аналитик данных') !=-1 or i['vactitle'].find('Data analyst') !=-1:
        l = i['vacdescription'].split(';')
        for _, j in enumerate(l):
            x = j.find('Python')
            if x != -1:
                b +=1
                break     
print(b)
```
